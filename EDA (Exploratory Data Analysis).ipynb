# Introduction of EDA:
Exploratory Data Analysis (EDA) is a part of Data Analysis (or Data Science) pipelines. As the name suggests, when we do exploration of the provided data to investigate and analyze it using high-level statistics, statistical graphs (or other data visualization methods) for summarizing its main characteristics then it is called Exploratory Data Analysis.


# Importance of EDA:
We use EDA to examine the data before making any assumptions and before using it for any predictive models. EDA helps us in understanding data, finding any outliers, constraints, and patterns in the data. With the help of EDA, we can validate our assumptions and we can ask the right questions about various statistics of features to get the answers we need.

### Relationship between means of Price where we have missing values and non missing values:

```python
for feature in features_with_na:
    data = df.copy() # creating a new dataset object with copy() method
    
    # let's make a variable that indicates 1 if the observation was missing or zero otherwise
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    
    # let's calculate the meadian price where the information is missing or present
    data.groupby(feature)['price'].median().plot.bar()
    plt.ylabel('price')
    plt.title(feature)
    plt.show()
```

## 10.1. Univariate Analysis:
Univariate analysis means “the analysis of one variable”. In this process analyze the density, distribution and find outliers for any variable.


### 10.1.1. Histogram: 
A histogram represents a frequency distribution by the means of rectangles whose widths represent class intervals and whose areas are proportional to the corresponding frequencies and the height of each rectangle is the average frequency density for the interval.


```python
# Histogram is a better way of representing 1D data than 1D Scatter plot
sns.histplot(df, x=df['price'])
```

Observation:
* Price follows Pareto like distribution. 
* It is similar to follow power law distribution.

### 10.1.2. PDF (Probability Density Function): 
The probability function representing the density of a Continuous Random Variable which is spread between a distinct range of values, is called Probability Density Function. 
In simple words, PDF represents the probability of the values of any continuous Random Variable.

### 10.1.3. CDF (Cumulative Distribution Function):
Cumulative sum of Probability Density Function is called Cumulative Distribution Function.


```python
# Plotting PDF and CDF

counts, bin_edges = np.histogram(df['price'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print("pdf: ",pdf);
print("\nbin_edges: ",bin_edges)

#compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf, label="pdf")
plt.plot(bin_edges[1:], cdf, label="cdf")
plt.legend()
plt.show()

# Disadvantage of PDF: Can we say what percentage of BMW Car points have a Price of less than 30,000?
# (Note: above disadvantage can be solved with CDF)

# Can we identify whether the PDF is Gaussian/ Normal or Uniform or Pareto or any other Distribution?
```

Observation:
* 90% Automobiles are having less than 25000 as price.

Since we can’t check 25th, 50th, and 75th percentile values through Probability Density Function therefore we use Box or Violin with whiskers plot.

### 10.1.4. Box and Whiskers Plot:

## 10.2. Bivariate Analysis:
Bivariate analysis is one of the statistical analyses where two variables are observed.

Graphs that are appropriate for bivariate analysis depend on the type of variable. 
For two continuous variables, a scatterplot is a common graph. 
When one variable is categorical and the other continuous, a box plot is common and 
when both are categorical a mosaic plot is common.


### Cases between two variables   and    Plot:
1. Both are continuous variables                           -   Scatter Plot
2. One variable is categorical and the other continuous    -   Bivariate Box Plot
3. Both are categorical variables                          -   Mosaic Plot

### 10.2.4. Pair Plot: 
With the help of a Pair Plot we can analyze both, the distribution of each individual numerical variable and the relationships between a pair of numerical variables. 

There is no limit for choosing the number of Numerical variables in Pair Plot, but it is more interpretable, useful and readable when we choose 3 to 6 features in the Pair Plot.

Note: We are using more than 2 variables in the Pair Plot, but the Pair Plot is a combination of Univariate and Bivariate Analyses. On the diagonal of the Plots matrix, we can find a Univariate analysis of the corresponding row **or** column feature (both are same for diagonal element). And, on any other place in the matrix, we can find the Bivariate analysis of the corresponding Row **and** Column feature.

```python
# Pairwise scatter plot: Pair-Plot
# Dis-advantages: 
## Can be used when number of features are high.
## Can not visualize higher dimensional patterns in 3-D and 4-D. 
## Only possible to view 2D patterns.


sns.pairplot(df[['length','horsepower','highway-mpg']]);
plt.show()

# NOTE: the diagnol elements are Histograms for each feature. 
```

Observation:
* There is a curvy relationship between horsepower and highway-mpg.
* Length and Highway-mpg are having Normal or Gaussian distribution.
* Horsepower has Right skewed distribution.

* Convertibles are having higher price with respect to any other body-styled cars (no matter what the length is).
* Wagons have medium to higher length with lower prices.

### 10.3.4. Heat Map:
In the Heat map we can analyse three variables in 2 Dimensional plot with the help of colour (as the magnitude of the third variable). The colour variation can be done by hue or intensity.  There are two different categories of heat maps: the cluster heat map and the spatial heat map. 

Note: 2D dataset that can be coerced into an ndarray for providing it to Heat Map. If a Pandas DataFrame is provided, the index/column information will be used to label the columns and rows.

Pivot Table and Correlation Matrix are ideal to create a Heat map.

## Heat Map of Correlation Matrix between independent features and dependent feature:


```python
# Correlation Matrix
cols = list(range(len(continuous_feature)+1)) # added 1 for Price (target) variable
fig, ax = plt.subplots(figsize=(15,15)) # defining the size and shape of the plot
corelation = df[continuous_feature+['price']].iloc[:,cols].corr() 
sns.heatmap(corelation, xticklabels = corelation.columns, yticklabels = corelation.columns, annot = True)
plt.title("Heat Map of Correlation Matrix between independent features and dependent feature",size=25)
plt.show()
```

Observations:
* Target value 'price' is positivly and highly correlated with width, curb-weight, engine-size, horsepower and city-L/100km features.
* Target value 'price' is negatively and highly correlated with city-mpg and highway-mpg.

### 10.3.5. Pairplot with Hue:
We can also use Pair Plot for Multivariate analysis with the help of hue. We can give any number of Continuous Numerical Variables on X and Y axes. And, then we can choose any Categorical Variable as Hue. Then we will get Kernal Density Estimation Plots on diagonal cells of the Pairplot and Scatter plot with hue on non-diagonal cells of the Pairplot.

```python
# Pair Plot with Hue
sns.pairplot(df[['length','horsepower','highway-mpg','diesel']], hue="diesel");
plt.show()
```

Observation:
* Curvy trend can be seen between horsepower and highway-mpg features.
* KDEs of Histograms are not well separated in length, horsepower and highway-mpg features.
* Diesel cars are at higher end of the length feature.
* Diesel cars are at lower end of the horsepower feature.
* Diesel cars are spreaded almost unifromly on highway-mpg feature.
