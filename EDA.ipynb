# Introduction of EDA:
Exploratory Data Analysis (EDA) is a part of Data Analysis (or Data Science) pipelines. As the name suggests, when we do exploration of the provided data to investigate and analyze it using high-level statistics, statistical graphs (or other data visualization methods) for summarizing its main characteristics then it is called Exploratory Data Analysis.


# Importance of EDA:
We use EDA to examine the data before making any assumptions and before using it for any predictive models. EDA helps us in understanding data, finding any outliers, constraints, and patterns in the data. With the help of EDA, we can validate our assumptions and we can ask the right questions about various statistics of features to get the answers we need.

12.   Correlation Matrix between independent features and dependent feature

# 5. Identifying the features:

## 5.1. Features with NaN or Null:

```python
## Here we will check the percentage of nan values present in each feature
## 1 -step make the list of features which has missing values
features_with_na = [features for features in df.columns if df[features].isnull().sum()>0]
```

```python
## 2- step print the feature name and the percentage of missing values

for feature in features_with_na:
    print(feature, np.round(df[feature].isnull().mean(), 4),  ' % missing values')
```

### If there are many missing values, then we can find the relationships between missing values and Target values

```python
for feature in features_with_na:
    data = df.copy() # creating a new dataset object with copy() method
    
    # let's make a variable that indicates 1 if the observation was missing or zero otherwise
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    
    # let's calculate the mean SalePrice where the information is missing or present
    data.groupby(feature)['price'].median().plot.bar()
    plt.ylabel('price')
    plt.title(feature)
    plt.show()
```

### Conclusion: 
Here, with the relation between the missing values and the dependent variable is clearly visible. So We need to replace these nan values with something meaningful which we will do in the Feature Engineering section.

# 8. Univariate, Bivariate, and Multivariate EDA:

To find the relationship between two or more Variables, we use the following plotting methods:
1. Univariate Analysis - PDF, CDF, Histogram, Box Plot, Violin Plot
2. Bivariate Analysis - 2D Scatter Plot, Bivariate Box Plot, Mosaic Plot, Pair Plot
3. Multi-Variate Analysis - Bar Plot, Heatmap, 3D Scatter Plot, 2D Scatter Plot with Hue


## 8.1. Univariate Analysis:
Univariate analysis means “the analysis of one variable”. In this process analyze the density, distribution and find outliers for any variable.


### 8.1.1. Histogram: 
It is used to summarize discrete or continuous data that are measured on an interval scale. In other words, a histogram represents a frequency distribution by the means of rectangles whose widths represent class intervals and whose areas are proportional to the corresponding frequencies: the height of each is the average frequency density for the interval.

We can choose which feature separates class labels in a better way:


```python
sns.histplot(df, x=df['price'])
```

### 8.1.2. PDF (Probability Density Function): 
The probability function representing the density of a Continuous Random Variable which is spread between a distinct range of values, is called Probability Density Function. 
In simple words, PDF represents the probability of the values of any continuous Random Variable.

### 8.1.3. CDF (Cumulative Distribution Function):
Cumulative sum of Probability Density Function is called Cumulative Distribution Function.


```python
#Plot CDF and pdf

counts, bin_edges = np.histogram(df['price'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print("pdf: ",pdf);
print("\nbin_edges: ",bin_edges)

#compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf, label="pdf")
plt.plot(bin_edges[1:], cdf, label="cdf")
plt.legend()
plt.show()
```

Since we can’t check 25th, 50th, and 75th percentile values through Probability Density Function therefore we use Box or Violin with whiskers plot.

### 8.1.4. Box and Whiskers Plot:
Box and whisker plot is a way to show the spread and quantiles of a random variable.

```python
sns.boxplot(x='price', data=df)
plt.show()
```

### 8.1.5. Violin and Whiskers Plot:
Violin plots are a method of plotting numeric data and can be considered a combination of the box plot with a kernel density plot.

```python
# A violin plot combines the benefits of the previous two plots 
#and simplifies them

# Denser regions of the data are fatter, and sparser ones thinner 
#in a violin plot

sns.violinplot(x="price", data=df, size=8)
plt.show()
```

## 8.2. Bivariate Analysis:
Bivariate analysis is one of the statistical analyses where two variables are observed.

Graphs that are appropriate for bivariate analysis depend on the type of variable. 
For two continuous variables, a scatterplot is a common graph. 
When one variable is categorical and the other continuous, a box plot is common and 
when both are categorical a mosaic plot is common.


### Case between two variables   and    Plot
1. Both are continuous variables                           -   Scatter Plot
2. One variable is categorical and the other continuous    -   Bivariate Box Plot
3. Both are categorical variables                          -   Mosaic Plot



### 8.2.1. Scatter Plot:
A scatter plot graphs the values of two variables against each other, and provides insight into their relationship.

```python
#2-D scatter plot:
# ALWAYS understand the axis: labels and scale.
df.plot(kind='scatter', x='price', y='horsepower') 
plt.show()
```

```python
vol = df.length * df.width * df.height
len(vol)
sns.scatterplot(data=df, x=df.price, y = vol)
plt.ylabel('Volume of vechicle')
plt.show()
```

### 8.2.2. Bivariate Box Plot:
When one variable is categorical and the other continuous then we can use Box plots for Bivariate Analysis


```python
sns.boxplot(x='price',y='body-style', data=df)
plt.show()
```

### 8.2.3. Mosaic Plot:
Mosaic plot is a graphical method for visualizing data from two or more categorical variables.
https://en.wikipedia.org/wiki/Mosaic_plot


```python
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic

# Create mosaic plot
mosaic(df,['aspiration', 'diesel'])
# Display the figure

plt.ylabel('Diesel')
plt.show()
```

Here in y axis 1 Meaning Diesel and 0 meand non Diesel (gas)

But if we want to Plot Bivariate analyses among many pairs of features along with Univariate analysis of each feature simultaneously, then we can use a Pair Plot.

### 8.2.4. Pair Plot: 
With the help of a Pair Plot we can analyze both the distribution of each variable and the relationships between two variables. There is no limit for choosing the number of variables in Pair Plot, but it is more interpretable, useful and readable when we choose 3 to 6 features in Pair Plot.

Note: We are using more than 2 variables in the Pair Plot, but the Pair Plot is a combination of Univariate and Bivariate Analyses. On the diagonal of the Plots matrix, we can find a Univariate analysis of the corresponding row **or** column feature. And, on any other place in the matrix, we can find the Bivariate analysis of the corresponding Row **and** Column.

```python
# pairwise scatter plot: Pair-Plot
# Dis-advantages: 
##Can be used when number of features are high.
##Cannot visualize higher dimensional patterns in 3-D and 4-D. 
#Only possible to view 2D patterns.
sns.set_style("whitegrid")
sns.pairplot(df[['length','horsepower','highway-mpg']]);
plt.show()
# NOTE: the diagnol elements are PDFs for each feature. PDFs are expalined below.
```

## 8.3. Multivariate Analysis:
Multivariate analysis is one of the statistical analyses where more than two variables are observed simultaneously.

### 8.3.1. Bar Plot with 3 Variables:
If we have two categorical variables and one numerical variable then we can plot it in a Bar Plot.

```python
sns.barplot(data=df, x='num-of-cylinders' ,y='price', hue='diesel')
plt.show()
```

### 8.3.2. Histogram with Hue:
Histogram with Hue as a Categorical Variable.

```python
sns.histplot(data=df, x = 'highway-mpg', hue= 'drive-wheels',kde=True)
```

### 8.3.3. Contour Plot:
A contour plot is a graphical technique for representing a 3-dimensional surface by plotting constant height (or z-axis) slices, called contours, on a 2-dimensional format. On a x,y 2D surface, we connect all the points with same z value.

```python
#2D Density plot, contors-plot
sns.jointplot(x="horsepower", y="highway-mpg", data=df, kind="kde");
plt.show();
```

### 8.3.4. Multivariate Scatter Plot:
Types: 
1. 3-dimensional Scatter Plot
2. 2-dimensional Scatter Plot with Hue

#### 8.3.4.1. 3D Scatter Plot:
A 3D Scatter Plot is three-dimensional plotting used to display the spread and properties of data as three variables of a dataset using the cartesian coordinates. To create a 3D Scatter plot, Matplotlib's mplot3d toolkit is used to enable three-dimensional plotting.

```python
# Import libraries
from mpl_toolkits import mplot3d
 
# Creating figure
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
 
# Creating plot
ax.scatter3D(df.length, df.width, df.height, color = "green")
plt.title("3D scatter plot")
 
# show plot
plt.show()
```

#### 8.3.4.2. 2D Scatter Plot with Hue: 
If we have 2 Numeric Variables and 1 Categorical Variable then we can use 2D Scatter Plot with Hue to analyse the relation among these 3 variables.  
Code- seaborn.scatterplot( x, y, data, hue)


```python
sns.scatterplot( data=df,x='length', y='price', hue='body-style')
```

### 8.3.5. Heat Map:
In the Heat map we can analyse three variables in 2 Dimensional plot with the help of colour (as the magnitude of the third variable). The colour variation can be done by hue or intensity.  There are two different categories of heat maps: the cluster heat map and the spatial heat map. 

Note: 2D dataset that can be coerced into an ndarray. If a Pandas DataFrame is provided, the index/column information will be used to label the columns and rows.

Pivot Tables and Correlation Matrix are ideal to create a Heat map.

## Heat Map of Correlation Matrix between independent features and dependent feature:


```python
# Correlation Matrix
cols = list(range(len(continuous_feature)+1)) # added 1 for Price (target) variable
fig, ax = plt.subplots(figsize=(15,15)) # defining the size and shape of the plot
corelation = df[continuous_feature+['price']].iloc[:,cols].corr() 
sns.heatmap(corelation, xticklabels = corelation.columns, yticklabels = corelation.columns, annot = True)
plt.title("Heat Map with Correlation Matrix between independent features and dependent feature",size=25)
plt.show()
```

### 8.3.6. Pairplot with Hue:
We can also use Pair Plot for Multivariate analysis with the help of hue. We can give any number of Continuous Numerical Variables on X and Y axes. And, then we can choose any Categorical Variable as Hue. Then we will get Kernal Density Estimation Plots on diagonal cells of the Pairplot and Scatter plot with hue on non diagonal cells of the Pairplot.

```python
# Pair Plot with Hue
sns.pairplot(df[['length','horsepower','highway-mpg','diesel']], hue="diesel");
plt.show()
```
